# Exercício 7 - Capítulo 13 

```python
import wooldridge
import statsmodels.formula.api as smf
import pandas as pd
```

```python
#Carregando a base de dados injury 
df = wooldridge.data('injury')
#Vendo a descrição do dataset
wooldridge.data('injury', description=True)
```

```python
#Filtrando o dataset para o Kentucky 
df_t = df[df.ky == 1] #1 é positivo para kentucky
```

```python
#Fazendo o modelo 
model = smf.ols('ldurat ~ afchnge + highearn + afhigh', data=df_t)
result = model.fit()
print(result.summary())
```

# Exercício C4 - Capítulo 13 

```python
#Usando o mesmo dataset do exercício anterior 
#Carregando a base de dados injury 
df = wooldridge.data('injury')
#Vendo a descrição do dataset
wooldridge.data('injury', description=True)
```

```python
#Filtrando o nosso dataset para o Kentucky 
df_t = df[df.ky == 1] #1 é positivo para kentucky
indust_dummy = pd.get_dummies(df_t['indust'], prefix='indust') #Criando dummies para industria
df_t = pd.concat([df_t, indust_dummy], axis=1) #Concatenando as dummies ao dataframe 
```

```python
#Renomeando as colunas igual ao R
df_t.rename(columns={"indust_1.0": "indust1", "indust_2.0": "indust2", "indust_3.0": "indust3"},inplace=True,)
df_t
```

```python
#Fazendo o modelo 
model = smf.ols('ldurat ~ afchnge + highearn + male + married + hosp + indust1 + indust2 + indust3 + afhigh', data=df_t)
result = model.fit()
print(result.summary())
```

```python
#Estimando para o Michigan 
df_m = df[df.mi == 1] #1 é positivo para Michigan
indust_dummy = pd.get_dummies(df_m['indust'], prefix='indust') #Criando dummies para industria
df_m = pd.concat([df_m, indust_dummy], axis=1) #Concatenando as dummies ao dataframe 
```

```python
#Renomeando as colunas igual ao R
df_m.rename(columns={"indust_1.0": "indust1", "indust_2.0": "indust2", "indust_3.0": "indust3"},inplace=True,)
df_m
```

```python
#Fazendo o modelo 
model = smf.ols('ldurat ~ afchnge + highearn + male + married + hosp + indust1 + indust2 + indust3 + afhigh', data=df_m)
result = model.fit()
print(result.summary())
```
